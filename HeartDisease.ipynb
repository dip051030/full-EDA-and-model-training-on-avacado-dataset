```python
num_columns = df.select_dtypes('number').columns

for col in num_columns:
    sns.histplot(data=df, x=col, hue="DEATH_EVENT", kde=True, bins=30)
    plt.title(col)
    plt.xlabel(f'mean: {round(df[col].mean(), 2)} median: {df[col].median()}')
    plt.show()
```

## Univariate Data Analysis

### 1. Age
Most people in the survey are between **40 and 70 years old**, and the distribution is **right-skewed**.
This likely happens because **heart-related problems are more common as people get older**, so middle-aged and elderly individuals are more represented in the dataset.

---

### 2. Creatinine Phosphokinase (CPK)
Creatinine phosphokinase is related to **quick energy use in muscles**, especially the heart.
The values are **not normally distributed** and show extreme spikes.
This suggests that **unusual energy stress in the body** may be linked to the **death event** in the dataset.

---

### 3. Diabetes
There are **more people without diabetes** than people with diabetes in the dataset.
This shows that diabetes is **not very common** in this group, though it can still increase health risk when combined with other conditions.

---

### 4. Ejection Fraction
Ejection fraction shows **how strongly the heart pumps blood**.
The normal range is **55% to 70%**, and higher values mean better heart performance.
This variable is **directly related to heart efficiency**, making it an important health indicator.

---

### 5. Platelets
Platelets are **tiny blood cells that help stop bleeding and repair the body**.
Based on the histogram and KDE plot, platelet values appear to be **normally distributed**, meaning most people fall within a healthy range.

---

### 6. Serum Creatinine
Serum creatinine is a **measure of how well the kidneys are cleaning waste from the blood**.
Higher values indicate weaker kidney function.
In this dataset, most people have **lower serum creatinine levels**, suggesting generally normal kidney health.

---

### 7. Serum Sodium
Serum sodium shows **how balanced the bodyâ€™s salt and water levels are**.
Low levels can cause weakness or confusion, while high levels suggest dehydration.
The distribution looks **close to normal**, with most people having healthy sodium levels.

---

### 8. Time
The `time` feature represents the **number of days each person was followed** in the study.
It helps understand **how long patients were observed**, which is important for outcome analysis.


```python
for col in num_columns:
    sns.boxplot(data=df, x="DEATH_EVENT", y=col)
    plt.title(col)
    plt.show()
```

## Bivariate Analysis (Feature vs Death Event)

### 1. Age vs Death Event
Older people tend to have **higher mortality** in this dataset.
The **median age** of people who died is around **65 years**, while for those who survived it is around **60 years**.
This means that, on average, people who died were **about 5 years older** than those who survived.
Also, age among the people who died shows **more variation**, indicating deaths are spread across a wider older age range.

---

### 2. Ejection Fraction vs Death Event
Ejection fraction shows a **strong relationship with mortality**.
The **median ejection fraction** of people who died is around **30**, while survivors have a higher median of about **37**.
Lower ejection fraction values are **more common among people who died**, showing that **weaker heart pumping ability increases the risk of death**.

---

### 3. Platelets vs Death Event
Platelet counts for people who died and those who survived are **very similar**.
The boxplots largely **overlap**, meaning platelet levels do not clearly separate the two groups.
This suggests that **platelet count has little or no strong association with mortality** in this dataset.

---

### 4. Serum Creatinine vs Death Event
Serum creatinine shows a **clear association with mortality**.
People who died generally have **higher serum creatinine levels** compared to those who survived.
This indicates that **poorer kidney function is linked to higher risk of death**.

---

### 5. Serum Sodium vs Death Event
Serum sodium has a **weak association with mortality**.
People who died tend to have **slightly lower sodium levels**, but the typical values for both groups mostly overlap.
This means sodium levels alone are **not a strong predictor of death**, though extreme values may still matter.

---

### 6. Time vs Death Event
The `time` feature shows a **very strong association with mortality**.
Most people who were followed for **longer periods survived**, while those with **shorter follow-up times died more often**.
This indicates that **patients who live longer naturally have longer follow-up durations**, making this feature highly informative for survival analysis.
