```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include=["object", "bool"]).columns

num_cols, cat_cols
```

```python
df["passed_exam"].value_counts()
```

```python
sns.countplot(x="passed_exam", data=df)
plt.title("Target Variable Distribution")
plt.show()
```

```python
df[num_cols].hist(figsize=(15, 10), bins=30)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[num_cols])
plt.xticks(rotation=90)
plt.show()
```

```python
for col in cat_cols:
    print(f"\n{col}")
    print(df[col].value_counts())
```

```python
sns.boxplot(x="passed_exam", y="hours_spent_learning_per_week", data=df)
plt.title("Study Hours vs Exam Outcome")
plt.show()
```

```python
important_features = [
    "practice_problems_solved",
    "projects_completed",
    "self_reported_confidence_python",
    "final_exam_score"
]

for col in important_features:
    sns.boxplot(x="passed_exam", y=col, data=df)
    plt.title(f"{col} vs Exam Outcome")
    plt.show()
```

```python
corr = df[num_cols].corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr, cmap="coolwarm", annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
corr["passed_exam"].sort_values(ascending=False)
```

```python
df.groupby("country")["passed_exam"].mean().sort_values(ascending=False)
```

```python
engagement_cols = [
    "uses_kaggle",
    "participates_in_discussion_forums"
]

for col in engagement_cols:
    sns.barplot(x=col, y="passed_exam", data=df)
    plt.title(f"{col} vs Exam Outcome")
    plt.show()
```
